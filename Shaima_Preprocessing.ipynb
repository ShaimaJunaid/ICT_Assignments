6. Do min max scaling on the feature set (Take ‘Survived’ as target)

```python
# Again load the same dataset
data=pd.read_csv("C:/DSA COURSE/DataSets/titanic_dataset.csv")
```

```python
data = data.set_index('PassengerId')
```

```python
data
```

```python
data.isnull().sum()
```

```python
# By using pandas to fill null values
# null values fill with median in case of numerical column
for i in[ "Age"]:
    data[i]=data[i].fillna(data[i].median())
# fill with most occuring value in case of categorical column     
for i in["Cabin","Embarked"]:
    data[i].fillna(data[i].value_counts().index[0],inplace=True)
```

```python
data.isnull().sum()
```

```python
# independent values
x = data.drop("Survived",axis=1)
x
```

```python
#Dependent value/Target
y = data["Survived"]
y
```

```python
# Encoding the categorical columns using label encoding
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
x["Name"] = label.fit_transform(x["Name"])
x["Sex"] = label.fit_transform(x["Sex"])
x["Ticket"] = label.fit_transform(x["Ticket"])
x["Cabin"] = label.fit_transform(x["Cabin"])
x["Embarked"] = label.fit_transform(x["Embarked"])
x
```

```python
# Dropping the columns which have less label values
x1 = x.drop(["Pclass","Sex","SibSp","Parch","Embarked"],axis=1)
x1
```

```python
x1.isnull().sum()
```

```python
# MinMax Scaling
from sklearn.preprocessing import MinMaxScaler
min_max = MinMaxScaler(feature_range =(0,1))
x1 = min_max.fit_transform(x1)
x1
```
