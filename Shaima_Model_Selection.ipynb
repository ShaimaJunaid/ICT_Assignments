2. Do all the necessary pre-processing steps

```python
#check the null values
data.isnull().sum()
```

```python
#calculate the percentage of null values
data.isnull().sum()*100/len(data)
```

Cabin's column null value is above 50% so drop that column

```python
#check the values
data["Cabin"].value_counts().unique
```

```python
# Drop Cabin
data=data.drop("Cabin",axis=1)
data
```

```python
#Check the value of embarked
data["Embarked"].value_counts()
```

```python
#check the value of age
data["Age"].value_counts()
```

```python
# null values can be fill with median in case of numerical column
for i in[ "Age"]:
    data[i]=data[i].fillna(data[i].median())
#filling "missing word" with categorical values
for i in[ "Embarked"]:
    data[i]=data[i].fillna("Missing")
```

```python
#Again check the null values
data.isnull().sum()
```

```python
data
```

no null values

```python
# drop some columns
data = data.drop(["Name","SibSp","Parch","Ticket"],axis=1)
data
```

```python
#Label Encoding Sex and Embarked
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
data["Sex"] = label.fit_transform(data["Sex"])
data["Embarked"] = label.fit_transform(data["Embarked"])
```

```python
data
```

```python
# Independent variables
x=data.drop(["Survived"],axis = 1)
x
```

```python
# Target variable
y=data["Survived"]
y
```

# Hold Out Technique

```python
# split the datas
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state = 42, test_size = 0.2)
```

3. Create kNN 

# KNN

```python
# Import KNeighborsClassifier
from sklearn.neighbors import KNeighborsClassifier
# Import confusion_matrix,accuracy_score
from sklearn.metrics import  accuracy_score
# assigning empty list
metric_k =[]
# kepping some values to neighbours 
neighbors = np.arange(3,15)
neighbors
```

```python
# create a classifier using euclidean
for k in neighbors:
    classifier = KNeighborsClassifier(n_neighbors = k,metric ="euclidean")
    classifier.fit(x_train,y_train)
    y_prediction = classifier.predict(x_test)
    acc = accuracy_score(y_test,y_prediction)
    metric_k.append(acc)
```

```python
# plot the k value
plt.plot(neighbors,metric_k,'o-')
plt.xlabel("K value")
plt.ylabel("Accuracy")
plt.grid()
```

```python
#K = 7 has the maximum value
classifier = KNeighborsClassifier(n_neighbors =7,metric ="euclidean")
classifier.fit(x_train,y_train)
y_prediction = classifier.predict(x_test)
```

```python
#Accuracy Score
print("Accuracy is " ,accuracy_score(y_test,y_prediction)*100 ,"%")
```

```python
#Accuracy check by using score function
score_KNN = classifier.score(x_test,y_test)
print("Accuracy (using score) is ",(score_KNN)*100 ,"%")
```

4. Do k-fold and stratified stratified k-fold cross validation techniques and find
the average accuracy score of the models

# K Fold Cross Validation

```python
#import kfold from sklearn
from sklearn.model_selection import KFold 
#Assign kfold as 10,default is 5
Kfold_validator = KFold(10) 
```

```python
# illustrate the training index and validation index using for loop
for train_index, test_index in Kfold_validator.split(x,y):
    print("TRAINING INDEX :\n ",train_index)
    print("\n---------------------------------------------------------------------------\n")
    print("VALIDATION INDEX :\n ",test_index)
    print("\n----------------------------------------------------------------------------\n")
```

# K Fold Cross Validation Score of KNN Model

```python
# import cross_val_score from sklearn
from sklearn.model_selection import cross_val_score
#find the score
kf_knn = cross_val_score(classifier,x,y,cv = Kfold_validator)
```

```python
#Cross validation Score of 10 fold
kf_knn
```

```python
# Mean of Cross validation Scores
cv_mean_knn =np.mean(kf_knn)
print("Cross validation Score is ",(cv_mean_knn)*100 ,"%")
```

# Stratified Cross Validation 

```python
# import StratifiedKFold from sklearn 
from sklearn.model_selection import StratifiedKFold 
#assign  n_split as 10
Sfold_validator = StratifiedKFold(n_splits =10)
```

```python
# illustrate the training index and validation index using for loop
for train_index, test_index in Sfold_validator.split(x,y):
    print("TRAINING INDEX :\n ",train_index)
    print("\n---------------------------------------------------------------------------\n")
    print("VALIDATION INDEX :\n ",test_index)
    print("\n----------------------------------------------------------------------------\n")
```

# Stratified Cross Validation Score of KNN

```python
#find the score
scv_result_knn = cross_val_score(classifier,x,y,cv = Sfold_validator)
```

```python
#Stratified Cross Validation score of 10 fold
scv_result_knn
```

```python
# Mean of Cross validation Scores
scv_mean_knn =np.mean(scv_result_knn)
print("Stratified Cross validation Score is ",(scv_mean_knn)*100 ,"%")
```
